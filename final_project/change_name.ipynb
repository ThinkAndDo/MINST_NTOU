<h1>Preprocess Image Data</h1>
<p>Insert file name and it will autoprocess file form that our model required.</p>
<p>step1: insert '<span style="color:red">man</span>' and press '<span style="color:red">enter</span>'<br>

   step3: re-run it and insert '<span style="color:red">woman</span>' and press '<span style="color:red">enter</span>'</p>

```python
import shutil
import os
import random
from os import walk
from os.path import join
import shutil

myPath = './res/people_face_picture'
print(myPath)
keyWord = input('the dir:')
input_path = join(myPath, keyWord)

if not os.path.exists(input_path):
            os.mkdir(myPath + keyWord)
            print('not', keyWord)



fileName = []

# get all image name from myPath
for root, dirs, files in walk(input_path):
    
    files.sort()
    
    if files:
        fileName = files

# print('RRR',root,'DDD',dirs,'FFF',files)


    
train_dir = myPath + "/train"
validation_dir = myPath + "/validation"
train_data_dir = train_dir +"/" + keyWord
validation_data_dir = validation_dir +"/" + keyWord

if not os.path.exists(train_dir):
    os.mkdir(train_dir)
if not os.path.exists(validation_dir):
    os.mkdir(validation_dir)
if not os.path.exists(train_data_dir):
    os.mkdir(train_data_dir)
if not os.path.exists(validation_data_dir):
    os.mkdir(validation_data_dir)



# get random 100 images from fileName and re-path them
# rng = [random.randint(0, len(fileName)) for _ in range(100)]
rng_list = []
while len(rng_list) < 100:
    rng = random.randint(0, len(fileName))
    
    if not rng in rng_list:
        rng_list.append(rng)
    else:
        continue
        
# print(len(rng_list), len(fileName))
    
rng_list.sort(reverse = True)

# add validation data
for i in rng_list:
    random_image_name = join(input_path, fileName[i])
#     print(random_image)
    shutil.copy(os.path.abspath(random_image_name), os.path.abspath(validation_data_dir))
    
#     FullPath = join(myPath, fileName[i]) #get all path
#     FileName = join(fileName[i]) #get file name
#     print(FullPath)
#     print(FileName)
#     print(keyWord)
    #shutil.move(FullPath, myPath + '/' + keyWord)
    fileName.pop(i)
print('success to copy file from', input_path, 'to', validation_data_dir)
    
    
# print(len(fileName))    

     
#add training data
for imgName in fileName:
    random_image_name = join(input_path, imgName)
    shutil.copy(os.path.abspath(random_image_name), os.path.abspath(train_data_dir))
print('success to copy file from', input_path, 'to', train_data_dir)
print('success copying file.')
    
    
```

<h1>Rename Image Data</h1>

```python
import os
path = './res/people_face_picture/'
path_train = './res/people_face_picture/train/'
path_validation = './res/people_face_picture/validation/'

list_root=[]
list_dirs=[]
list_files=[]

# first, rename training data
# get training folder information
for root, dirs, files in walk(path_train):
    list_root.append(root)
    list_dirs.append(dirs)
    list_files.append(files)

# start rename
for i in range(len(list_root)-1):
    n = 1;
    for imgName in list_files[i+1]:
        oldName = list_root[i+1] + "/" + imgName
        if n<10:
            newName = list_root[i+1] + "/" + list_dirs[0][i] + '00' + str(n) + '.png'
        elif n<100:
            newName = list_root[i+1] + "/" + list_dirs[0][i] + '0' + str(n) + '.png'
        else:
            newName = list_root[i+1] + "/" + list_dirs[0][i] + str(n) + '.png'
        os.rename(oldName, newName)
        print(oldName, '=====>', newName)
        n += 1


# then, rename validation data
# get training folder information
list_root.clear()
list_dirs.clear()
list_files.clear()
for root, dirs, files in walk(path_validation):
    list_root.append(root)
    list_dirs.append(dirs)
    list_files.append(files)

# start rename
for i in range(len(list_root)-1):
    n = 1;
    for imgName in list_files[i+1]:
        oldName = list_root[i+1] + "/" + imgName
        if n<10:
            newName = list_root[i+1] + "/" + list_dirs[0][i] + '00' + str(n) + '.png'
        elif n<100:
            newName = list_root[i+1] + "/" + list_dirs[0][i] + '0' + str(n) + '.png'
        else:
            newName = list_root[i+1] + "/" + list_dirs[0][i] + str(n) + '.png'
        os.rename(oldName, newName)
        print(oldName, '=====>', newName)
        n += 1
```
